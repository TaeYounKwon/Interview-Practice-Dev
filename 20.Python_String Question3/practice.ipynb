```python
def solution(logs):
    answer = dict()
    
    id = []
    
    # O(n)
    for info in logs:
        if info[1] not in id:
            id.append(info[1])
    
    minmax = [[10000000000000,-1] for i in range(len(id))]
    
    # O(n)
    for info in logs:
        if info[1] in id:
            numb = int(info[0])
            # find the index
            findVal = id.index(info[1])
            
            # update the min-max values according to index
            if minmax[findVal][0]>numb:
                minmax[findVal][0]=numb
                
            if minmax[findVal][1]<numb:
                minmax[findVal][1]=numb
                
    for k in range(len(id)):
        answer[id[k]] = minmax[k]
         
                
    
    return answer
```

```python
print('Result 1: ',solution(logs1))
print('Result 2: ',solution(logs2))
print('Result 3: ',solution(logs3))
print('Result 4: ',solution(logs4))
```
